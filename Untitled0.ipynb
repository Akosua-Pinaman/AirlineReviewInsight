```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.airlinequality.com/airline-reviews/british-airways"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)

if response.status_code == 400:
    print("Page fetched successfully!")
else:
    print("Failed to fetch page, status code: {response.status_code}")
```

```python

soup = BeautifulSoup(response.text, "html.parser")
```

```python
reviews = soup.find_all("div",
class_ = "text_content")

for index, review in enumerate(reviews[:20]):
    print("Review", str(index + 1), ":", review.text.strip(), "\n")
```

```python
import csv
```

```python
with open("british_airways_reviews.csv", "w", newline="", encoding = "utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Reviews"])

    for review in reviews:

      writer.writerow([review.text.strip()])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("british_airways_reviews.csv")
df.head()
```

```python
import re
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)


    return text

df["Cleaned_Review"] = df["Reviews"].apply(clean_text)

df.head()
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
df["Cleaned_Review"].head(10)
```

```python
df = df.dropna(subset = ["Cleaned_Review"])
df["Cleaned_Review"] = df["Cleaned_Review"].astype(str)
```

```python
sample_text = "".join(df["Cleaned_Review"].sample(10))


```

```python
print(df.columns)
```

```python
wordcloud = WordCloud(width=800, height=400, max_words= 100, background_color="white").generate(sample_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("Off")
plt.title("Most Common words in the review")

plt.show()
```
